**Solución del exámen:**

Importar las librerias necesarias, crear el Dataset y leer los datos:

```python
import pandas as pd
import numpy as np

exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Crear el DataFrame
df = pd.DataFrame(exam_data, index=labels)
df
```

1. Seleccionar solo las columnas name y score:

```python
columnas_seleccionadas = df[['name', 'score']]
columnas_seleccionadas
```

2. Filtrar las filas donde el número de intentos (attemps) sea mayor que 2:

```python
Filas_filtradas_attempts = df[df['attempts'] > 2]
Filas_filtradas_attempts
```

3. Filtrar las filas donde el score se encuentre entre 15 y 20 (incluidos):

```python
filas_filtradas_score = df[(df['score'] >= 15) & (df['score'] <= 20)]
filas_filtradas_score
```

4. Cambiar el score en la fila 'd' a 11.5:

```python
cambio_score = df.loc['d', 'score'] = 11.5
df
```

5. Agregar una nueva fila con etiqueta 'k' al dataframe con un valor deseado para cada columna:



```python
df.loc['k'] = ['John Byron', 18.7, 2, 'yes']
df
```

6. Borrar la fila 'k' y entregar nuevamente el DataFrame original:

```python
df = df.drop('k', axis=0)
df
```

7. Ordene el DataFrame por nombre en orden ascendente:

```python
df_ascendente = df.sort_values(by='name', ascending=True)
df_ascendente
```

8. Reemplazar los valores de la columna qualify que contienen los valores yes y no por los valores booleanos True y False respectivamente:

```python
df['qualify'] = df['qualify'].replace({'yes': True, 'no': False})
df
```

## **Ejercicio 2** (30%)

Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes
2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

5. ¿Cuál es el pais con más participación femenina?

```python
# punto 1 del Ejercicio 2

# importar librerias
import numpy as np
import pandas as pd

#crear DataFrame a apartir de los datos athletes.csv
athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")

#mosotras los primeros 10 registros del DataFrame
athletes.head(10)
```

```python
# punto 2 del Ejercicio 2

# Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.

# Se crea un dataframe nuevo athletes_ganadores, usando el idexador "loc" y el eperador "or" se filtran los
# atletas que han ganado al menos una medalla.

athletes_ganadores = athletes.loc[(athletes.gold >= 1) | (athletes['silver'] >= 1) | (athletes['bronze'] >= 1)]

# mostrar el dataframe athletes_ganadores

athletes_ganadores

#print(athletes_ganadores)
```

```python
# punto 3 del Ejercicio 2

# Calcule la masa muscular por fila con una función

# importar librerias
import numpy as np
import pandas as pd

def calculo_imm(data_frame):

    """
    La funcion recibe un dataframe y crea una nueva columna llamada Indice de
    masa muscular (Imm), donde el Imm es el resultado de dividir el peso (weight)
    entre la estatura (height) elevada al cuadrado.

    Ejemplo:

    data = {
        'sport': ['soccer', 'basketball', 'tennis'],
        'weight': [50, 60, 73],
        'height': [1.9, 1.7, 1.78]
    }

    In [1]:calculo_imm(data)
    Out[1]:

           sport   weight   eight   Imm
    0     soccer   50      1.90  13.850416
    1	  basket   60      1.70  20.761246
    3     tennis   73      1.78  23.040020

    """

    data_frame["Imm"] = data_frame['weight'] / data_frame['height']**2

    return data_frame

# llamar la funcion

calculo_imm(athletes)

# usar ayuda con la funcion
# help(calculo_imm)
```

```python
# punto 4 del Ejercicio 2

# Agrupe los datos calculando la media por deporte para cada una del resto de características.
# ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?


# elimino la columna "id" ya que no es necesario un promedio de las "id"

sin_id = athletes.drop(['id'], axis=1)

# se crea un dataframe que muestra la media por deporte para cada característica

media_deporte = sin_id.groupby('sport').agg([np.mean])

# muestra el dataframe

media_deporte
```

```python
# muestra el dataframe media_peso con cada deporte como indice y el promedio del peso de mayor a menor

media_peso = sin_id.groupby('sport').agg({'weight': np.mean}).sort_values(by='weight', ascending=False)

# muestra el dataframe media_peso del promedio del peso de forma descendente para cada deporte

media_peso

#print(media_peso)
```

```python
# punto 5 del Ejercicio 2

# ¿Cuál es el pais con más participación femenina?

# se cuentan los valores unicos de la caracteristica nacionalidad en el
# dataframe "athletes" filtrado por genero femenino para mostrar el país
# con mayor participacion.

s = athletes.loc[athletes['sex'] == "female", "nationality"].value_counts()

# se crea un data frame a partir de la serie creada con la condicion anterior

dfs = s.to_frame()

# se muestra el pais con más participación femenina

dfs[:1]
```

## **Ejercicio 3 (30%)**

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información. Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta (mediante un código de color) la posición.
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
#Ejercicio 3

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Se importa el dataset nba del repositorio respectivo:
data = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv")
dff = pd.DataFrame(data)
#Elimino la columna Unnamed: 0 para una mejor visualización
dff = dff.drop(columns=['Unnamed: 0'])
dff
#Se eliminan las instancias con registros faltantes
df = dff.dropna()
df
```

```python
#PUNTO 1
#1. Cuáles son los equipos que pagan los salarios más altos
#(presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)


#RESPUESTA:
#Se calcula el total del salario por equipo, y de esos se tienen en cuenta los cinco con los salarios más altos
ds = df.groupby('Team').Salary.agg(['sum']).sort_values(by = 'sum', ascending = False)
dt = ds.head(5)

#Se renombra la columna 'sum' por 'Total_Salario' para mejor visualización
new_column_names = {
    'sum': 'Total_Salario'
}

#Finalmente, se muestran los cinco equipos con la nómina más cara, además del total de la nómina
dt = dt.rename(columns=new_column_names)
dt
```

```python
#PUNTO 2
#2. Separe la variable Age en tres grupos y cree una nueva columna llamada Age-Group con esta información.
#Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta
#(mediante un código de color) la posición.

#RESPUESTA:

#Se crean los rangos de edad y las etiquetas respectivas
age_bins = [19, 26, 36, 45]
age_labels = ['19 a 25 años', '26 a 35 años', '36 a 45 años']

#Se añade la nueva columna ('Age-Group) a la base trabajada
df['Age-Group'] = pd.cut(df['Age'], bins=age_bins, labels = age_labels, right=False)
df
```

```python
#Se cuentan los jugadores según el grupo de edad al que pertenece, y la posición del jugador
datos_agrupados = df.groupby(['Age-Group', 'Position']).size().reset_index(name='Count')
datos_agrupados
```

```python
#Se crea un índice que trae los tres grupos de edades
indice_grupo_edad = datos_agrupados['Age-Group'].unique()
print(indice_grupo_edad)
```

```python

#Se traen también los índices o arreglos que contienen los jugadores
#según su rango de edad y posición

print("Numero de jugadores por edad [19 a 25 años, 26 a 35 años, 36 a 45 años")

datos_posicion_c = datos_agrupados[datos_agrupados['Position'] == 'C']
datos_posicion_c2 = datos_posicion_c['Count'].to_numpy()
print(f"Posición C: {datos_posicion_c2}")

datos_posicion_pf = datos_agrupados[datos_agrupados['Position'] == 'PF']
datos_posicion_pf2 = datos_posicion_pf['Count'].to_numpy()
print(f"Posición PF: {datos_posicion_pf2}")

datos_posicion_pg = datos_agrupados[datos_agrupados['Position'] == 'PG']
datos_posicion_pg2 = datos_posicion_pg['Count'].to_numpy()
print(f"Posición PG: {datos_posicion_pg2}")

datos_posicion_sf= datos_agrupados[datos_agrupados['Position'] == 'SF']
datos_posicion_sf2= datos_posicion_sf['Count'].to_numpy()
print(f"Posición SF: {datos_posicion_sf2}")

datos_posicion_sg = datos_agrupados[datos_agrupados['Position'] == 'SG']
datos_posicion_sg2 = datos_posicion_sg['Count'].to_numpy()
print(f"Posición SG: {datos_posicion_sg2}")
```

```python
#Finalmente, se crea la gráfica teniendo como etiquetas los rangos de edad
#Y como rango, el número de jugadores
#además de eso, se muestra por posición en el campo:

resultado = pd.DataFrame({'C' : datos_posicion_c2,
                     'PF' : datos_posicion_pf2,
                     'PG' : datos_posicion_pg2,
                     'SF' : datos_posicion_sf2,
                     'SG' : datos_posicion_sg2,
                    },

                    index=(indice_grupo_edad))

#Se configuran finalmente las características
#de la tabla (ancho, etiquetas, rangos, largo de los ejes, etc....)
n = len(resultado.index)
x = np.arange(n)
width = 0.10

plt.bar(x - width, resultado.C, width=width, label='C')
plt.bar(x, resultado.PF, width=width, label='PF')
plt.bar(x + width, resultado.PG, width=width, label='PG')
plt.bar(x + 2*width, resultado.SF, width=width, label='SF')
plt.bar(x + 3*width, resultado.SG, width=width, label='SG')
plt.ylabel("Número de jugadores")
plt.xlabel("Grupo de edad")
plt.title("Número de jugadores por grupo de edad y posición")
plt.xticks(x, resultado.index)
plt.legend(loc='best')
plt.show()
```

```python
#PUNTO 3
#3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición
# para los cinco equipos que pagan los mejores salarios.


#RESPUESTA:

#Creo una variable que siempre referencie a los índides de los cinco equipos con la mayor nómina
#(esto ayuda a que si la base inicial es alimentada o alterada, no sea necesario cambiar este código
# y por el contrario, se actualice y siga mostrando los cinco equipos con la mejor nómina)
indice_columna = dt.index.unique()

#Se crea la tabla dinámica especificando el salario promedio de cada uno de los cinco equipos,
#separados por posición
equipos_incluir = indice_columna
df_filtrado = df[df["Team"].isin(equipos_incluir)]
td_final = df_filtrado.pivot_table("Salary",index="Position",columns="Team")

#Finalmente, se muestra la tabla dinámica, y para una mejor visualización
#se redondea el resultado a dos decimales
td_final_rounded = td_final.round(2)
td_final_rounded
```
