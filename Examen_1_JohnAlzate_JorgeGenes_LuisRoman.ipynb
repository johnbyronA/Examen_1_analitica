```python
# Se genera la clase Pedido, correspondiente al pedido de compra realizado a través de una
# entidad bancaria, la cual posee los siguientes atributos:

# ID: id de la transacción
# full_name: nombre completo del cliente
# ammount: cantidad de la transaccion por el pedido
# purchase_type: tipo de transacción, internacional (INT) o nacional (NAL)


class Pedido():
  def __init__(self,  ID: int, full_name: str,ammount: float, purchase_type:str):
    self.info={'ID': ID,
               'full_name':full_name,
               'ammount': ammount,
               'purchase_type':purchase_type
               }

# Se define método id_persona para mostrar información general del movimiento del cliente con el pedido

  def id_persona(self):
    id_persona=self.info['ID']
    auto_name=self.info['full_name']
    print(f"El nombre del cliente es {auto_name} y el ID de su pedido es {id_persona}")

# Se define un método cobrar_iva, el cual calculará el IVA asociado al monto de la transacción (19% de la misma)

  def cobrar_iva(self):
    iva=self.info['ammount']*0.19
    print(f"El IVA asociado al monto de la transacción es ${iva} COP")

# Se define una método tipo_compra, el cual determinará a cuantas cuotas se diferirá el pago dependiendo si la transacción es nacional o internacional.

  def tipo_compra(self):
    if self.info['purchase_type']=="NAL":
      print("La compra es nacional y se diferirá automáticamente a una cuota")
    elif self.info['purchase_type']=="INT":
      print("La compra es internacional y se diferirá automáticamente a 24 cuotas")
    else:
      print("El dato de la transacción es erróneo, reintentar ingreso de datos")


# Clase hija:
#

class PedidoDescuento(Pedido):
    def __init__(self, ID: int, full_name: str, ammount: float, purchase_type: str, descuento: float):
        super().__init__(ID, full_name, ammount, purchase_type)
        self.descuento = descuento

# Se crea un método aplicar_descuento, el cual generará un descuento del 20% a la transacción si esta supera el monto de los $150000 COP

    def aplicar_descuento(self):
        #monto_con_descuento=(self)
        if self.info['ammount'] > 150000:
          monto_con_descuento=self.info['ammount']*0.2
          print(f"El monto de compra es de ${self.info['ammount']} COP y es aplicable al descuento del 20% del monto, con el descuento aplicado de ${monto_con_descuento} la transacción queda en ${self.info['ammount'] - monto_con_descuento} COP")
        else:
          print(f"El monto no es lo suficientemente alto para obtener el descuento")

# Crear una instancia de la clase PedidoDescuento
pedido_con_descuento = PedidoDescuento(198765, "Ana Rodríguez", 164567, "NAL", 0.15)

# Llamar a los métodos heredados de la clase pedido
pedido_con_descuento.id_persona()
pedido_con_descuento.cobrar_iva()
pedido_con_descuento.tipo_compra()

# Llamar al método específico de PedidoDescuento
pedido_con_descuento.aplicar_descuento()

#jorge.info

#Andres=Pedido(12345,"Jorge Luis Genes Padilla", 23500.50,"NAL").cobrar_iva()





```

```python
#Sea el siguiente ecuaciones lineales de 5*5:

#              2x  + 3y  -  z  + 4w  + 5v =   15
#               x  - 2y  + 4z  - 3w  +  v =    6
#              3x  + 2y  - 3z  + 5w  - 2v =   11
#              4x  +  y  - 2z  + 3w  + 2v =    8
#               x  +  y  +  z  +  w  +  v =    7

#Para resolverlo, se hace lo siguiente:

#Se importa la Biblioteca Numpy y se renombra como 'np'
import numpy as np

#Se crea un array con los coeficientes de cada variable (x,y, z, w, v)
x = np.array([[2, 3, -1, 4, 5],
              [1, -2, 4, -3, 1],
              [3, 2, -3, 5, -2],
              [4, 1, -2, 3, 2],
              [1, 1, 1, 1, 1]])

#Se crea un array con las soluciones esperadas de cada ecuación
y = np.array([15, 6, 11, 8, 7])

#Se utiliza la función linalg.solve de la biblioteca Numpy
resultado = np.linalg.solve(x,y)


#El resultado será los valores de las variables x, y, z, w, v que dan solución al Sistema de Ecuaciones.
print(f"La solución al sistema de ecuaciones lineales anterior se compone de:")

print(f" x = {resultado[0]}, \n y = {resultado[1]}, \n z = {resultado[2]}, \n w = {resultado[3]}, \n v = {resultado[4]}")
```

```python
#Sea la ecuación de Euler(e) igual a: la sumatoria desde 0 hasta un número m de la función 1/n!
#e = (1/0! + 1/1! + 1/2! +  1/3! + ...)
#Teniendo en cuenta que 1/0! = 1
#También Euler(e) se puede expresar como: 1 más la sumatoria desde 1 hasta un número m de la función 1/n!
#e = (1 + 1/1! + 1/2! + 1/3! + ...)


#Se importa la Biblioteca Numpy y se renombra como 'np'
import numpy as np


#Se crea una función que devuelve el valor de Euler(e) al hacer la sumatoria hasta un número específico m
def Py_numpy(m):

#Se crea un array desde 1 hasta m+1
 n = np.arange(1,m+1).astype(float)

#Se crea una variable b que equivale a la fórmula para calcular el valor de Euler(e)
 b = 1+np.sum(1/np.cumprod(n))
 return b

#Me muestra el valor calculado.
valor_calc_euler = Py_numpy(8)
valor_calc_euler_round = round(Py_numpy(8),5)

print(f"El valor calculado de Euler(e) es ≈ {valor_calc_euler} ≈ {valor_calc_euler_round}")

#Ahora, necesitamos saber el número mínimo de términos para garantizar una
#precisión de 5 decimales.

# Para que el número calculado respete los 5 decimales, la diferencia entre el número real de Euler(e)
#y el número aproximado debe ser máximo de 0.00001

dif_minima = 0.00001

# Se crea una función para que calcule la diferencia entre el valor aproximado y el valor real de Euler(e),
#y evalúe qué valor mínimo de m cumple con la garantización de los 5 decimales.

#NOTA: Tener en cuenta que el número de términos totales utilizados para calcular Euler (e) es m + 1,
#ya que es necesario tener en cuenta el valor cuando m = 0.
#Explicado mediante un ejemplo, si se requiere calcular el valor de Euler sumando los valores desde m = 0 hasta 6,
#realmente estoy utilizando 7 términos para llegar a ese valor (0,1,2,3,4,5,6).

def evaluar_diferencia(m, dif_minima):

    dif_valores = abs(Py_numpy(m) - np.e)
    if dif_valores <= dif_minima:
        return m + 1
    else:
        return evaluar_diferencia(m + 2, dif_minima)

#Se inicia con m = 0 ya que es el primer valor de m.
m = 0

# Finalmente, el código nos arroja el número de terminos que se requiere para que el número de Euler(e) calculado
#garantice los primeros 5 decimales respecto al número de Euler(e) real:

no_terminos = evaluar_diferencia(m, dif_minima)

print(f"Se necesitan {no_terminos} términos para garantizar una exactitud de 5 decimales.")
```

```python
import numpy as np

def Polinomio_caracteristico(A):

    """
    La funcion recibe una matriz n x n

    Sea A una matriz cuadrada n x n, al escalar lambda: λ se le llama eigenvalor
    de la matriz A si existe un vector X distinto de cero tal que AX = λX.
    Tal vector X se llama eigenvector de A correspondiente a λ. El polinomio
    caracteristico esta dado por p(λ) = det(A-λI), donde det es el determinante
    e I es la matriz identidad, es decir que en la diagonal principal todos son
    unos y el resto son ceros

    Ex:
    La matriz A:
    A = np.array ([[3,  2], [7,  -2]])

    tiene los siguientes Autovalores (eigenvalor)
    Autovalores: 5.0, -4.0

    Y su polinomio caracteristico esta dado por:
    (λ - 5) (λ + 4)

     In[0]: Polinomio_caracteristico(A)
     Out[0]: El Polinomio Característico es: (λ - (5.0)) (λ - (-4.0))
    """

    # Calcular el polinomio característico utilizando la libreria numpy
    polynomial_coefficients = np.poly(A)

    # Obtener los autovalores a partir de los coeficientes del polinomio
    autovalores = np.roots(polynomial_coefficients)

    # Redondear los autovalores para mostrarlos de forma más legible
    autovalores_redondeados = np.round(autovalores, decimals=2)

    # Convertir los autovalores en una lista de cadenas de texto
    autovalores_str = [f"{val.real} + {val.imag}i" if val.imag != 0 else f"{val.real}" for val in autovalores_redondeados]

    # Imprimir el polinomio caracteristico de la matriz (n x n)
    solucion = "El Polinomio Característico es: (λ - (" + ")) (λ - (".join(autovalores_str) + "))"

    return solucion

# Matriz de ejemplo 1
A = np.array([[2, 1, 1], [4, 3, 2], [1, 1, 2]])

# Matriz de ejemplo 2
#A = np.array ([[3,  2], [7,  -2]])

# mostrar el  polinomio
solucion = Polinomio_caracteristico(A)
print(solucion)
```
